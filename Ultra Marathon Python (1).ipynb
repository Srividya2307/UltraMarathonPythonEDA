```python
#import libraries
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
#See the data thats imported
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

```python
#clean up data
#Only retrieve USA races, 50km or 50 miles, for year 2020 
```

```python
df[df['Event distance/length']=='50km']
```

```python
#Combining 50km and 50 miles with isin
df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
#Filtering with only 2020 data
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020)]
```

```python
#Filtering with races that took place in the USA
df[df['Event name'].str.endswith('(USA)', na=False)]
```

```python
#Combining all the three conditions
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020) & (df['Event name'].str.endswith('(USA)', na=False))]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
#Clean up athlete age 
df2['Athlete_Age'] = 2020-df2['Athlete year of birth']
```

```python
#Remove h from athlete performance
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(5)
```

```python
#Drop columns - Athlete Club, Athlete Country, Athlete year of birth, Athlete age category
df2=df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis=1)
```

```python
df2.head()
```

```python
#Clean up null values
df2.isna().sum()
```

```python
df2[df2['Athlete_Age'].isna()==1]
```

```python
df2=df2.dropna()
```

```python
df2.shape
```

```python
#Check for duplicates
df2[df2.duplicated()==True]
```

```python
#reset index
df2.reset_index(drop=True)
```

```python
#Fix datatypes
```

```python
df2.dtypes
```

```python
df2['Athlete_Age']=df2['Athlete_Age'].astype(int)
```

```python
df2['Athlete average speed']=df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
#Rename Columns
```

```python
#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#Athlete_Age                    int32
```

```python
df2 = df2.rename(columns = {'Year of event': 'year',
                            'Event dates': 'race_date',
                            'Event name': 'race_name',
                            'Event distance/length': 'race_length',
                            'Event number of finishers': 'race_no_of_finishers',
                            'Athlete performance': 'athlete_performance',
                            'Athlete gender': 'athlete_gender',
                            'Athlete average speed': 'athlete_avg_speed',
                            'Athlete ID': 'athlete_id',
                            'Athlete_Age': 'athlete_age',   
                               
})
```

```python
df2.head()
```

```python
#Reorder columns
```

```python
df3 = df2[['race_date','race_name','race_length','race_no_of_finishers','athlete_id','athlete_performance','athlete_avg_speed','athlete_gender','athlete_age']]
```

```python
df3.head()
```

```python
sns.histplot(df3['race_length'])
```

```python
sns.histplot(df3,x='race_length',hue='athlete_gender')
```

```python
sns.displot(df3[df3['race_length']=='50mi']['athlete_avg_speed'])
```

```python
sns.violinplot(data = df3, x='race_length', y='athlete_avg_speed',hue='athlete_gender',split='True',inner='quarts')
```

```python
sns.lmplot(data=df3, x= 'athlete_age', y='athlete_avg_speed', hue='athlete_gender')
```

```python
#Difference in speed for the 50km, 50mi male to female
```

```python
df3.groupby(['race_length','athlete_gender'])['athlete_avg_speed'].mean()
```

```python
#What age groups are the best in the 50mi Race (20+ races min.)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean',ascending=False).query('count>19')
```

```python
#What age groups are slow in the 50mi Race (20+ races min.)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean',ascending=True).query('count>19')
```

```python
#Seasons for the data --> Slower in summer than in winter?

#Spring 3-5
#Summer 6-8
#Fall 9-11
#Winter 12-2

#Split between two decimals
```

```python
df3['race_month']=df3['race_date'].str[3:5].astype(int)
```

```python
df3.head()
```

```python
df3['race_season']=df3['race_month'].apply(lambda x: 'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>2 else 'Winter')
```

```python
df3.head(20)
```

```python
df3.groupby('race_season')['athlete_avg_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```
